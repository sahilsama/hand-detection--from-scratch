```python
import mediapipe as mp
import cv2 #for our webcam
import numpy as np 
import uuid
import os
```

```python
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands
# mp_face = mp.solutions.face
```

```python
import cv2
cap = cv2.VideoCapture(0)

with mp_hands.Hands(min_detection_confidence = 0.8, min_tracking_confidence=0.5) as hands: 
    while cap.isOpened():
    #     frame represent our img from webcam
        ret, frame = cap.read()
        
        
        #detection
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        #set flags
        image.flags.writeable = False
    
    
#         print("Frame Shape:", frame.shape)  # Check if frame is captured correctly

        #detetions
        results = hands.process(image)
        
   
        
#       set flag tp truw
        image.flags.writeable = True
    
#       set bgr to rgb
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
#       detections  
        print(results)
    
        if results.multi_hand_landmarks:
            for num, hand in enumerate(results.multi_hand_landmarks):
                mp_drawing.draw_landmarks(image,hand, mp_hands.HAND_CONNECTIONS,
                        mp_drawing.DrawingSpec(color=(121,20,76), thickness=2, circle_radius=4),
                        mp_drawing.DrawingSpec(color=(121,44,76), thickness=2, circle_radius=2),
                        )
            
        
        
        cv2.imshow('Handtracking', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    
cap.release()
cv2.destroyAllWindows()
```
